```python
import itertools as iter

for item in iter.chain([1,2],[1,2]):# разбирает цепь на составляющие
    print(item)
```

```python
i =0
for item in iter.cycle([1, 2]):
    i = i +1
    print(item)
    if i >10:
        break
```
